```python
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime


def scrape_amazon_laptops():
    URL = "https://www.amazon.in/s?k=laptop"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.5993.90 Safari/537.36"
    }

    response = requests.get(URL, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    laptops = []


    results = soup.find_all('div', {'data-component-type': 's-search-result'})

    for result in results:

        ad_label = result.find('span', string="Sponsored")
        is_ad = "Ad" if ad_label else "Organic"


        title_tag = result.h2
        title = title_tag.text.strip() if title_tag else "No Title"

        image_tag = result.find('img', {'class': 's-image'})
        image_url = image_tag['src'] if image_tag else "No Image"


        rating_tag = result.find('span', {'class': 'a-icon-alt'})
        rating = rating_tag.text.split()[0] if rating_tag else "No Rating"


        whole_price = result.find('span', {'class': 'a-price-whole'})
        fraction_price = result.find('span', {'class': 'a-price-fraction'})
        if whole_price and fraction_price:
            price = whole_price.text + fraction_price.text
        else:
            price = "No Price"

        laptops.append([image_url, title, rating, price, is_ad])


    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    filename = f"sujithreddy_amazon_laptops_{timestamp}.csv"
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Image URL', 'Title', 'Rating', 'Price', 'Ad/Organic'])
        writer.writerows(laptops)

    print(f"Scraping completed. Data saved to {filename}")

if __name__ == "__main__":
    scrape_amazon_laptops()
```
